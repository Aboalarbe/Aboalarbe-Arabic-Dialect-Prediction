### Import Libraries

```python
import requests
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

### Load data

```python
df = pd.read_csv('data/dialect_dataset.csv')
print(len(df))
display(df.head())
```

### Preprocessing

```python
df['id'] = df['id'].astype(str)
id_list = df['id'].to_list()
```

### Split IDs into batchs of 1000

```python
def id_chunks_1000(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
request_lists = list(id_chunks_1000(id_list,1000))
print(len(request_lists))
print(len(request_lists[-2]))
print(len(request_lists[-1]))
```

### Send POST Request

```python
txt_response = []
for lst in request_lists:
    r = requests.post("https://recruitment.aimtechnologies.co/ai-tasks", json= lst)
    print(r.status_code, r.reason)
    json_obj = r.json()
    for id in lst:
        response = json_obj[id]
        txt_response.append(response)
```

### Create Dataframe with response data

```python
data = pd.DataFrame(list(zip(id_list, txt_response, df['dialect'])), columns =['id', 'txt','dialect'])
print(len(data))
display(data.head())
```

### Save dataframe into Excel

```python
data.to_excel('data/data.xlsx')
print('Done!')
```
